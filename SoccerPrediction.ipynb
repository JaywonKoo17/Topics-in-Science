```
import numpy as np
import pandas as pd
```

```
game_data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/epl2020.csv")
```

```
print(game_data.shape)
```

```
POSITIONS = np.array(range(1, 21))
# Get the list of teams
teams = game_data['teamId'].unique()

# Get the results for each team
team_results = []
for team in teams:
    # Get the data for that team
    team_data = game_data[game_data['teamId'] == team]
    
    wins = team_data['wins'].sum()
    draws = team_data['draws'].sum()
    losses = team_data['loses'].sum()
    scored = team_data['scored'].sum()
    conceded = team_data['missed'].sum()
    games = wins + draws + losses
    points = (3 * wins) + draws
    goal_difference = scored - conceded
    
    team_results.append([team, games, wins, draws, losses, scored, conceded, goal_difference, points])

league_table = pd.DataFrame(team_results, columns=['Team', 'P', 'W', 'D', 'L', 'F', 'A', 'GD', 'Points'])
league_table.sort_values(by=['Points', 'GD', 'F'], ascending=False, inplace=True, ignore_index=True)
league_table.set_index(POSITIONS, inplace=True)
league_table.head(20)
```

```
teams = game_data['teamId'].unique()

each_teamResults = []
for team in teams:
    # Get the data for that team
    team_data = game_data[game_data['teamId'] == team]
    result_seq = team_data['result'].values
    each_teamResults.append([team, result_seq.tolist()]) 
    print([team, result_seq.tolist()])
#print(each_teamResults)
```

```
# For overall percentage 
#print(len(each_teamResults[0][1]))
win_total = 0
lose_total = 0
draw_total = 0
ww = 0
wl = 0
wd = 0
lw = 0
ll = 0
ld = 0
dw = 0
dd = 0
dl = 0

for team in each_teamResults:
  for game in range(1, len(team[1])):
    before = team[1][game-1]
    now = team[1][game]
    if before == 'w':
      win_total += 1
      if now == 'w':
        ww += 1
      elif now == 'l':
        wl += 1
      else:
        wd += 1
    elif before == 'l':
      lose_total += 1
      if now == 'w':
        lw += 1
      elif now == 'l':
        ll += 1
      else:
        ld += 1
    else:
      draw_total += 1
      if now == 'w':
        dw += 1
      elif now == 'l':
        dl += 1
      else:
        dd += 1

print("=============Results=============")
print("Total Games:", win_total+lose_total+draw_total)
print("Win:", win_total)
print("ww:",ww,"wl:",wl,"wd:",wd)
print("--> [Percentage] ww:",ww/win_total,"wl:",wl/win_total, "wd:",wd/win_total ) 
print("") 
# 0.42 0.35 0.23
print("Lose:", lose_total)
print("lw:",lw,"ll:",ll,"ld:",ld,"total:", lw+ll+ld)
print("--> [Percentage] ww:",lw/lose_total,"ll:",ll/lose_total, "ld:",ld/lose_total )
# 0.34 0.39 0.27
print("")

print('Draw:', draw_total)
print("dw:",dw,"wl:",dl,"wd:",dd,"total:", dw+dl+dd)
print("--> [Percentage] ww:",dw/draw_total,"wl:",dl/draw_total, "wd:",dd/draw_total ) 
# 0.36 0.38 0.26
print("")  
```
